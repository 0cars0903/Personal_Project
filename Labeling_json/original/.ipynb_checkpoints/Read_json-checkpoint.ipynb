```python
import json

import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Box 형태 라벨링
def box(class_id, coord, color):
    x, y, w, h = coord
    cv2.putText(cvt_img, class_id, (int(x+0.5*w), int(y-15)), cv2.FONT_HERSHEY_SIMPLEX,1.0,color,3)
    cv2.rectangle(cvt_img, (int(x), int(y)), (int(x+w), int(y+h)), color, 5) #(x,y) 좌표
    
# 다각형 라벨링
def polylabel(poly):
    x, y = 0, 0
    l = len(poly)
    for lst in poly:
        x += lst[0]
        y += lst[1]
    x = x / l
    y = y / l
    return int(x), int(y)

def polygon(class_id, coord, color):
    points = np.array(coord, np.int32)
    a = polylabel(coord)
    cv2.putText(cvt_img, class_id, (a[0], a[1]), cv2.FONT_HERSHEY_SIMPLEX,1.0,color,3)
    cv2.polylines(cvt_img, [points], True, color, 3) #(x,y) 좌표

# 색상 출력
import random

def rand_color():
    col = [i for i in range(0, 256)]
    return random.choice(col)

# Data Load
def load_data(DIR_img, DIR_label, name):
    # 이미지 불러오기
    img_name = DIR_img + name + ".jpg"
    img = cv2.imread(img_name)
    cvt_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # annotation 불러오기
    json_name = DIR_label + name + '.json'
    a_json = open(json_name, encoding = 'utf-8')
    a_dict = json.load(a_json)
    label = a_dict["Learning data info."]["annotation"]
    return cvt_img, label

# 라벨링 프로그램
def labeling(label):
    # label 형태에 따라 실행 
    for dic in label:
        class_id = dic['class_id']
        type_label = dic['type']
        coord = dic['coord']
        if type_label == 'box':
            color = (rand_color(),rand_color(),rand_color())
            box(class_id, coord, color)
        else:
            color = (rand_color(),rand_color(),rand_color())
            polygon(class_id, coord, color)
            
# 출력 및 저장
def print_img(cvt_img):
    plt.imshow(cvt_img)
    cvt_img = cv2.cvtColor(cvt_img, cv2.COLOR_RGB2BGR)
    cv2.imwrite(name+'_label'+'.jpg', cvt_img)
```

```python
DIR_img = "C:/Users/piai/Desktop/work/test/original/"
DIR_label = "C:/Users/piai/Desktop/work/test/original/"
name = "L-210929_G06_J_UC-13_002_0034"
cvt_img, label = load_data(DIR_img, DIR_label, name)
labeling(label)
print_img(cvt_img)
```
